```python
def blur_faces(image_path):
    # Load pre-trained face detection model
    face_cascade = cv.CascadeClassifier('/content/drive/MyDrive/PCVK/Images/haarcascades/haarcascade_frontalface_alt.xml')

    # Load the image
    image = cv.imread(image_path)

    # Convert the image to grayscale
    gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Blur faces
    for (x, y, w, h) in faces:
        face = image[y:y+h, x:x+w]
        face = cv.GaussianBlur(face, (99, 99), 5)  # Adjust the kernel size for stronger or weaker blurring
        image[y:y+face.shape[0], x:x+face.shape[1]] = face

    # Display the result
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

image_path = '/content/drive/MyDrive/PCVK/Images/facedet/mjordan.jpg'
blur_faces(image_path)
```

```python
import os
from tqdm import tqdm
import cv2
import random
import numpy as np
import pickle


# Training dataset directory
DATADIR = "/content/drive/MyDrive/PCVK/Images/dataset/training"
dirs = []  # List of character names (categories)
training_data = []  # List to store training data
width, height = 100, 100  # Resize dimensions

# Loop through the dataset directories
for char_name in sorted(os.listdir(DATADIR)):
    dirs.append(char_name)

# Loop through images in each directory
for char_name in dirs:
    path = os.path.join(DATADIR, char_name)
    class_number = dirs.index(char_name)  # Assign a numeric label to each category
    for img in tqdm(os.listdir(path), desc=f"Processing {char_name}"):
        try:
            # Read image
            img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
            # Resize to the target size
            new_array = cv2.resize(img_array, (width, height))
            # Append the data with its label
            training_data.append([new_array, class_number])
        except Exception as e:
            pass

# Shuffle the data to ensure randomness
random.shuffle(training_data)

# Split features and labels
X = []
Y = []
for feature, label in training_data:
    X.append(feature)
    Y.append(label)

# Convert to numpy arrays
X = np.array(X).reshape(-1, width, height, 1)  # Add channel dimension
Y = np.array(Y)

# Save data to pickle files
with open("/content/X.pickle", "wb") as pickle_out:
    pickle.dump(X, pickle_out)
with open("/content/Y.pickle", "wb") as pickle_out:
    pickle.dump(Y, pickle_out)
```
